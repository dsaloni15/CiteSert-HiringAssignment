```python
#Importing necessary libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def get_path(directory,columns):
    dataframes = []
    for f in os.listdir(directory):
        if f.endswith('.asc'):
            Company_Name = os.path.splitext(f)[0][:-2]
            filepath = os.path.join(directory,f)
            df = pd.read_csv(filepath,header=None,names=columns)
            df['Company_Name'] = Company_Name
            dataframes.append(df)
    return pd.concat(dataframes,ignore_index=True)
```

```python
def convert_to_minute(df):
    df['Date'] = df['Date'].astype(str)
    df['Time'] = df['Time'].astype(str)

    df['Datetime'] = pd.to_datetime(df['Date']+' '+df['Time'],format='%m/%d/%Y %H:%M:%S.%f')
    df['Minute'] = df['Datetime'].dt.floor('min')

    return df

trade_data = convert_to_minute(trade_data)
quote_data = convert_to_minute(quote_data)
```

```python
# Group by 'Minute' and 'Company_Name'
trade_agg = trade_data.groupby(['Minute', 'Company_Name']).agg({
    'Trade Id': 'count',          # Number of trades
    'Price': ['first', 'max', 'min', 'last'],  # Open, High, Low, Close
    'Volume': 'sum',              # Total volume
    'Buy Order Capacity': 'sum',  # Total buy capacity
    'Sell Order Capacity': 'sum', # Total sell capacity
    'Trade Period': lambda x: x.mode()[0],  # Mode of Trade Period
})

# Flatten column names
trade_agg.columns = ['num_trades', 'Open', 'High', 'Low', 'Close', 'total_volume',
                     'total_buy_cap', 'total_sell_cap', 'trade_period']

# Reset index to turn it back into a DataFrame
trade_agg = trade_agg.reset_index()

# Compute Volume-Weighted Average Price (VWAP)
trade_agg['weighted_price'] = trade_data.groupby(['Minute', 'Company_Name']).apply(
    lambda x: np.average(x['Price'], weights=x['Volume'])
).values

# Compute Trade Imbalance Ratios
trade_agg['trade_imbalance_ratio'] = trade_data.groupby(['Minute', 'Company_Name'])['Aggressor Side'].apply(
    lambda x: (x == 'B').sum() / (x == 'S').sum() if (x == 'S').sum() > 0 else 0
).values

trade_agg['trade_volume_imbalance_ratio'] = trade_data.groupby(['Minute', 'Company_Name'])['Volume'].apply(
    lambda x: trade_data.loc[x.index, 'Volume'][trade_data.loc[x.index, 'Aggressor Side'] == 'B'].sum() /
              trade_data.loc[x.index, 'Volume'][trade_data.loc[x.index, 'Aggressor Side'] == 'S'].sum()
              if trade_data.loc[x.index, 'Volume'][trade_data.loc[x.index, 'Aggressor Side'] == 'S'].sum() > 0 else 0
).values

trade_agg['order_cap_imbalance_ratio'] = trade_data.groupby(['Minute', 'Company_Name'])['Sell Order Capacity'].apply(
    lambda x: (x.isna().sum()) / len(x)
).values
```

```python
# Compute the spread once to avoid repetitive calculations
quote_data['spread'] = quote_data['Ask Price'] - quote_data['Bid Price']

# Group by 'Company_Name' and 'Minute'
quote_agg = quote_data.groupby(['Company_Name', 'Minute']).agg({
    'spread': ['mean', 'max', 'min'],   # Average, max, min spread
    'Bid Size': 'sum',                  # Total bid size
    'Ask Size': 'sum',                  # Total ask size
})

# Rename the columns for clarity
quote_agg.columns = ['avg_spread', 'max_spread', 'min_spread', 
                     'total_bid_size', 'total_ask_size']

# Reset index to make it a regular DataFrame
quote_agg = quote_agg.reset_index()

# Compute Weighted Average Prices
quote_agg['weighted_avg_bid_price'] = quote_data.groupby(['Company_Name', 'Minute']).apply(
    lambda x: np.average(x['Bid Price'], weights=x['Bid Size']) if x['Bid Size'].sum() > 0 else np.nan
).values

quote_agg['weighted_avg_ask_price'] = quote_data.groupby(['Company_Name', 'Minute']).apply(
    lambda x: np.average(x['Ask Price'], weights=x['Ask Size']) if x['Ask Size'].sum() > 0 else np.nan
).values
```

```python
df = pd.merge(trade_agg,quote_agg,on=['Company_Name','Minute'],how='outer')
df
```

```python
df.dropna(inplace=True)
df
```

```python
#Percentage change calculation
df['Next Close'] = df.groupby('Company_Name')['Close'].shift(-1)
df['Return'] = (df['Next Close'] - df['Close']) / df['Close']
df.dropna(subset=['Next Close', 'Return'], inplace=True)
df['Company_Name'].unique()
df
```

```python
#Train-Test Split(80-20)
train_data,test_data = pd.DataFrame(),pd.DataFrame()

for company in df['Company_Name'].unique():
    company_df = df[df['Company_Name']==company].sort_values('Minute')
    split = int(0.8*len(company_df))

    train_data = pd.concat([train_data,company_df.iloc[:split]])
    test_data = pd.concat([test_data,company_df.iloc[split:]])
```

```python
# Step 6: Model Training

features = ['num_trades', 'Open', 'High', 'Low', 'Close', 'total_volume', 
            'total_buy_cap', 'total_sell_cap', 'trade_imbalance_ratio', 
            'trade_volume_imbalance_ratio', 'order_cap_imbalance_ratio', 
            'avg_spread', 'max_spread', 'min_spread', 'total_bid_size', 
            'total_ask_size', 'weighted_avg_bid_price', 'weighted_avg_ask_price']

model = LinearRegression()

# Fit the model for each Company
results = {}
for company in train_data['Company_Name'].unique():
    company_train_data = train_data[train_data['Company_Name'] == company]
    model.fit(company_train_data[features], company_train_data['Return'])
    # Store the model or results
    results[company] = model

# Predictions
test_data['Predicted Return'] = np.nan
for company in test_data['Company_Name'].unique():
    company_test_data = test_data[test_data['Company_Name'] == company]
    test_data.loc[test_data['Company_Name'] == company, 'Predicted Return'] = model.predict(company_test_data[features])
```
